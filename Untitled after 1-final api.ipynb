```python
import pandas as pd
import numpy as np

# --- 1. Load Data and Create Deep Copy ---

# NOTE: Replace 'original_data_file.csv' with actual file name
file_name = 'data1.csv'
# Assuming still need to skip the first 56 rows based on earlier context
df_original = pd.read_csv(file_name, skiprows=56) 

# IMPORTANT: Create the deep copy for all cleaning operations
df_working = df_original.copy() 

# Identify numerical columns for both Imputation and Outlier Removal
numerical_cols = df_working.select_dtypes(include=np.number).columns.tolist()

print("--- Starting Data Cleaning on Copy ---")
print(f"Original number of rows: {len(df_working)}")
initial_missing = df_working.isnull().sum().sum()
print(f"Total missing values before cleaning: {initial_missing}")

# --- 2. Median Imputation (Fill Null Values) ---

cols_to_impute = df_working.select_dtypes(include=np.number).columns[
    df_working.select_dtypes(include=np.number).isnull().any()
].tolist()

if cols_to_impute:
    for col in cols_to_impute:
        # Calculate the median and fill the NaN values
        median_value = df_working[col].median()
        df_working[col].fillna(median_value, inplace=True)
    print("\n Step 1: Median Imputation complete for numerical columns.")
else:
    print("\nStep 1: No numerical columns needed Median Imputation.")


# --- 3. Outlier Handling (Capping/Winsorization) ---

def cap_iqr_outliers(df, column):
    """Caps outliers in a specific column to the IQR bounds instead of removing rows."""
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # We use 1.5 for standard outliers, or 3.0 for extreme ones
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Use numpy to cap values: anything below lower becomes lower, anything above upper becomes upper
    df[column] = np.clip(df[column], lower_bound, upper_bound)
    
    return df

rows_before = len(df_working)

# Apply Capping to all numerical columns
for col in numerical_cols:
    df_working = cap_iqr_outliers(df_working, col)

rows_after = len(df_working)

print(f"Step 2: Outlier Capping complete.")
print(f"Rows preserved: {rows_after} (Out of {rows_before})")


# --- Step 1.5: Categorical Imputation ---

# Identify columns that contain text/objects
categorical_cols = df_working.select_dtypes(include=['object', 'category']).columns

# Fill all missing values in these columns with "Unknown"
df_working[categorical_cols] = df_working[categorical_cols].fillna("Unknown")

print(f"Step 1.5: Categorical Imputation complete. Text columns filled with 'Unknown'.")

# --- 4. Final Output and Save ---

# Define the new file name as requested
output_file_name = 'new_correct.csv' 
df_working.to_csv(output_file_name, index=False)

final_missing = df_working.isnull().sum().sum()

print("\n--- Final Confirmation ---")
print(f"Final number of rows: {len(df_working)}")
print(f"Final total missing cells (should be 0 for numerical columns): {final_missing}")
print(f"The cleaned dataset has been saved to: {output_file_name}")
```

```python
# --- 1. Load the Data (Fixing the NameError) ---
file_name = "new_correct.csv"
df = pd.read_csv(file_name) 

# --- 2. Feature Engineering: Habitability Score Index ---
EARTH_RADIUS = 1.0
EARTH_EQT = 288.0

# Calculate sub-scores using numpy's abs for vectorized operation
df['Size_Score'] = 1 / (1 + np.abs(df['pl_rade'] - EARTH_RADIUS))
df['Temp_Score'] = 1 / (1 + np.abs(df['pl_eqt'] - EARTH_EQT))

# Calculate final index and drop intermediate columns
df['Habitability_Score_Index'] = (df['Size_Score'] + df['Temp_Score'])
df.drop(columns=['Size_Score', 'Temp_Score'], inplace=True, errors='ignore') # Use errors='ignore' in case columns are already dropped

# --- 3. Feature Engineering: Insulation Factor ---
df['Insulation_Factor'] = df['st_teff'] / df['pl_eqt']

# --- 4. Update the Original File ---
df.to_csv(file_name, index=False)

print(f"Code executed successfully. The file '{file_name}' has been updated with new features.")
print(f"The updated DataFrame now has {df.shape[1]} columns.")
```

```python
T_SUN = 5778.0  # Sun's effective temperature in Kelvin
MET_SUN = 0.0   # Sun's metallicity ([Fe/H])
LUM_SUN = 0.0   # Sun's luminosity (st_lum is in log(Solar Luminosity))

df['Stellar_Temp_Score'] = 1 / (1 + np.abs(df['st_teff'] - T_SUN) / T_SUN)
df['Stellar_Met_Score'] = 1 / (1 + np.abs(df['st_met'] - MET_SUN))
df['Stellar_Lum_Score'] = 1 / (1 + np.abs(df['st_lum'] - LUM_SUN))

# Calculate final index and drop intermediate columns
df['Stellar_Compatibility_Index'] = (
    df['Stellar_Temp_Score'] + 
    df['Stellar_Met_Score'] + 
    df['Stellar_Lum_Score']
) / 3

df.drop(columns=['Stellar_Temp_Score', 'Stellar_Met_Score', 'Stellar_Lum_Score'], inplace=True, errors='ignore')

print(" Stellar Compatibility Index created.")


# --- Verification Print (This is the section that prints the results) ---
print("\n--- Final Results (First 5 Rows) ---")
print(f"Final DataFrame now has {df.shape[1]} columns.")
print(df[['Stellar_Compatibility_Index']].head())
```

```python
import pandas as pd

# --- Step 1: Targeting specific file ---
target_file = 'final_processed_data1.csv'
df_encoded = pd.read_csv(target_file)

# --- Step 2: Define Targets ---
df_encoded['is_habitable'] = (df_encoded['Habitability_Score_Index_Scaled'] > 0.8).astype(int)
target_score = 'Habitability_Score_Index_Scaled'

# --- Step 3: Select Features via Correlation ---
# Use numeric_only=True to ignore text columns like planet names
corr_matrix = df_encoded.corr(numeric_only=True)[target_score].abs()

# Filter for correlation > 0.1 and drop targets from the feature list
selected_features = corr_matrix[corr_matrix > 0.1].drop([target_score, 'is_habitable'], errors='ignore').index.tolist()

# --- Step 4: Verification Prints ---
print(f"Targets Defined: {df_encoded['is_habitable'].sum()} habitable planets identified.")
print(f"Feature Selection: {len(selected_features)} features selected based on correlation > 0.1.")
print(f"Top Correlated Features: {selected_features[:5]}")
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score, precision_score
import joblib
import numpy as np

# --- 1. Data Selection ---
leakers = ['pl_rade_Scaled', 'pl_radj']
final_features = [f for f in selected_features if f not in leakers]
X = df_encoded[final_features]
y = df_encoded['is_habitable']

# --- 2. THE MODEL ---
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier(
        n_estimators=200,
        max_depth=4,                 
        min_samples_leaf=200,        
        max_features=0.5,            # Limit features per tree
        class_weight={0: 1, 1: 5.0}, 
        random_state=42
    ))
])

# --- 3. Cross-Validation ---
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_accuracies = cross_val_score(pipeline, X, y, cv=cv, scoring='accuracy')

# --- 4. Final Evaluation (Test Set) ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
y_probs = pipeline.predict_proba(X_test)[:, 1]

# --- 5. Results ---
print("="*45)
print("5-FOLD CROSS-VALIDATION (ACCURACY)")
print("="*45)
for i, score in enumerate(cv_accuracies, 1):
    print(f"Fold {i}: {score:.4f}")
print("-" * 45)
print(f"MEAN CV ACCURACY: {np.mean(cv_accuracies):.4f}")
print("="*45)

print("="*45)
print(f"Model Saved As: habitable_model.pkl")
print("="*45)

joblib.dump(pipeline, 'habitable_model.pkl')
```

```python
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
import numpy as np

# --- 1. Train Logistic Regression ---
lr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced'))
])
lr_pipeline.fit(X_train, y_train)

# --- 2. Train XGBoost (Cleaned & Regularized) ---
xgb_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', XGBClassifier(
        random_state=42,
        max_depth=3,               
        n_estimators=100,
        learning_rate=0.05,
        reg_lambda=10,             
        reg_alpha=5,               
        scale_pos_weight=12,       
        eval_metric='logloss'
    ))
])
xgb_pipeline.fit(X_train, y_train)

# --- 3. Get Predictions from all models ---
y_pred_rf = pipeline.predict(X_test)   #existing Random Forest
y_pred_lr = lr_pipeline.predict(X_test) 
y_pred_xgb = xgb_pipeline.predict(X_test)

# --- 4. Extract Metrics for the Report ---
rf_report = classification_report(y_test, y_pred_rf, output_dict=True)
lr_report = classification_report(y_test, y_pred_lr, output_dict=True)
xgb_report = classification_report(y_test, y_pred_xgb, output_dict=True)

comparison_data = {
    "Metric": ["Accuracy", "Recall (Class 1)", "Precision (Class 1)"],
    "Random Forest": [
        f"{accuracy_score(y_test, y_pred_rf):.4f}",
        f"{rf_report['1']['recall']:.4f}",
        f"{rf_report['1']['precision']:.4f}"
    ],
    "Logistic Regression": [
        f"{accuracy_score(y_test, y_pred_lr):.4f}",
        f"{lr_report['1']['recall']:.4f}",
        f"{lr_report['1']['precision']:.4f}"
    ],
    "XGBoost": [
        f"{accuracy_score(y_test, y_pred_xgb):.4f}",
        f"{xgb_report['1']['recall']:.4f}",
        f"{xgb_report['1']['precision']:.4f}"
    ]
}

df_compare = pd.DataFrame(comparison_data)

import joblib

# Save the entire pipeline (scaler + XGBoost model)
joblib.dump(xgb_pipeline, 'habitable_model_xgboost.joblib')

print("Model saved successfully as: habitable_model_xgboost.joblib")

# --- 5. Final Display ---
print("="*80)
print("FINAL MODEL COMPARISON: RF vs LR vs XGBOOST")
print("="*80)
print(df_compare.to_string(index=False))
print("="*80)
```

```python
from flask import Flask, request, jsonify, send_file
import joblib
import pandas as pd
import numpy as np
import io
from werkzeug.serving import run_simple

app = Flask(__name__)

# ---------------- LOAD MODEL ----------------
model = joblib.load("habitable_model_xgboost.joblib")
print("MODEL LOADED")

EXPECTED_COLUMNS = list(model.feature_names_in_)
print("EXPECTED FEATURES:", EXPECTED_COLUMNS)

# ---------------- LOAD TRAINING DATA FOR CORRELATIONS ----------------
train_df = pd.read_csv("final_processed_data1.csv")  # Replace with your CSV
train_numeric = train_df[EXPECTED_COLUMNS]  # only model features

# ---------------- PREDICTION ROUTE ----------------
@app.route("/predict", methods=["POST"])
def predict():
    try:
        data = request.get_json()

        # --- USER INPUT ---
        pl_bmasse = float(data["planet_mass_earth"])
        pl_orbper = float(data["orbital_period_days"])
        sy_dist   = float(data["orbit_distance_au"])
        st_teff   = float(data["star_temperature_k"])
        pl_rade   = float(data["star_radius_solar"])

        # --- BASE FEATURES ---
        base_features = {
            "pl_bmasse_Scaled": pl_bmasse,
            "pl_orbper_Scaled": pl_orbper,
            "sy_dist_Scaled": sy_dist,
            "st_teff_Scaled": st_teff,
            "pl_rade_Scaled": pl_rade
        }

        # --- ENGINEERED FEATURES ---
        engineered_features = {
    "Insulation_Factor_Scaled": st_teff / (sy_dist + 1e-6),
    "pl_bmassprov_Unknown": 0,
    "pl_radeerr1": pl_rade * 0.05,
    "pl_radeerr2": pl_rade * 0.02,
    "pl_radjerr1": pl_rade * 0.01,
    "pl_radjerr2": pl_rade * 0.03,
    "st_tefferr1": st_teff * 0.01,
    "st_tefferr2": st_teff * 0.02,
    "sy_disterr1": sy_dist * 0.05,
    "sy_disterr2": sy_dist * 0.02,
    # ----- ADD MISSING FEATURES -----
    "st_met_Scaled": 1.0,
    "st_meterr1": 0.0,
    "st_meterr2": 0.0,
    "soltype_Published Confirmed": 1,
    "soltype_TESS Project Candidate": 0
}


        # --- MERGE FEATURES ---
        full_input = {**base_features, **engineered_features}
        df = pd.DataFrame([full_input])
        df = df.reindex(columns=EXPECTED_COLUMNS, fill_value=0)

        # --- MODEL PREDICTION ---
        prob = model.predict_proba(df)[0][1]
        prediction = int(prob >= 0.45)

        # --- FEATURE IMPORTANCE (handle pipeline) ---
        try:
            if hasattr(model, 'named_steps'):
                last_step = list(model.named_steps.values())[-1]
                fi = last_step.feature_importances_
            else:
                fi = model.feature_importances_
        except:
            fi = [0]*len(EXPECTED_COLUMNS)

        main_features = ['pl_bmasse_Scaled','pl_orbper_Scaled','sy_dist_Scaled','st_teff_Scaled','pl_rade_Scaled']
        fi_values = [float(dict(zip(EXPECTED_COLUMNS, fi)).get(f,0)) for f in main_features]

        # --- HABITABILITY SCORE DISTRIBUTION ---
        all_scores = list(np.random.normal(loc=prob, scale=0.1, size=50))
        all_scores = [float(min(max(s,0),1)) for s in all_scores]

        # --- CORRELATION MATRIX ---
        correlations = train_numeric.corr().round(2).values.tolist()
        correlations = [[float(c) for c in row] for row in correlations]

        # --- RETURN JSON ---
        return jsonify({
            "habitability_score": float(round(prob,4)),
            "habitable": int(prediction),
            "feature_importance": fi_values,
            "all_scores": all_scores,
            "correlations": correlations
        })

    except Exception as e:
        return jsonify({"error": str(e)}), 400


# ---------------- EXPORT ANALYSIS TO EXCEL ----------------
@app.route("/export_csv", methods=["POST"])
def export_csv():
    try:
        data = request.get_json()
        fi = data["feature_importance"]
        scores = data["all_scores"]
        correlations = data["correlations"]

        df_fi = pd.DataFrame({
            "Feature": ['Planet Mass','Orbital Period','Orbit Distance','Star Temp','Star Radius'],
            "Importance": fi
        })
        df_scores = pd.DataFrame({"Habitability_Score": scores})
        df_corr = pd.DataFrame(correlations, columns=['Mass','Period','Distance','Temp','Radius'],
                               index=['Mass','Period','Distance','Temp','Radius'])

        output = io.BytesIO()
        with pd.ExcelWriter(output, engine='xlsxwriter') as writer:
            df_fi.to_excel(writer, sheet_name="Feature Importance", index=False)
            df_scores.to_excel(writer, sheet_name="Score Distribution", index=False)
            df_corr.to_excel(writer, sheet_name="Correlation Matrix")
        output.seek(0)
        return send_file(output, download_name="planet_analysis.xlsx", as_attachment=True)
    except Exception as e:
        return jsonify({"error": str(e)}), 400

# ---------------- EXPORT ANALYSIS TO PDF ----------------
@app.route("/export_pdf", methods=["POST"])
def export_pdf():
    try:
        from matplotlib.backends.backend_pdf import PdfPages
        import matplotlib.pyplot as plt
        import seaborn as sns

        data = request.get_json()
        fi = data["feature_importance"]
        scores = data["all_scores"]
        correlations = data["correlations"]

        output = io.BytesIO()
        with PdfPages(output) as pdf:
            # Feature Importance
            plt.figure(figsize=(6,4))
            plt.bar(['Planet Mass','Orbital Period','Orbit Distance','Star Temp','Star Radius'],
                    fi, color=['#6366f1','#22d3ee','#4ade80','#fbbf24','#f472b6'])
            plt.title("Feature Importance")
            pdf.savefig()
            plt.close()

            # Score Distribution
            plt.figure(figsize=(6,4))
            plt.hist(scores, bins=15, color='#6366f1', edgecolor='black')
            plt.title("Habitability Score Distribution")
            pdf.savefig()
            plt.close()

            # Correlation Matrix
            plt.figure(figsize=(6,4))
            sns.heatmap(correlations, annot=True, xticklabels=['Mass','Period','Distance','Temp','Radius'],
                        yticklabels=['Mass','Period','Distance','Temp','Radius'], cmap="viridis")
            plt.title("Correlation Matrix")
            pdf.savefig()
            plt.close()

        output.seek(0)
        return send_file(output, download_name="planet_analysis.pdf", as_attachment=True)

    except Exception as e:
        return jsonify({"error": str(e)}), 400

# ---------------- MAIN ----------------
if __name__ == "__main__":
    run_simple("127.0.0.1", 5000, app, use_reloader=False)
```
