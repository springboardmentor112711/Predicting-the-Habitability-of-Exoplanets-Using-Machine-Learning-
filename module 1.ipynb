```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("exoplanet.csv", low_memory=False)
df.shape
```

```python
df.head()
```

```python
df.info
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df.drop_duplicates(subset=["pl_name"], inplace=True)
df.shape
```

```python
columns_needed = [
    "pl_name", "pl_orbper", "pl_rade", "pl_bmasse", "pl_dens",
    "st_teff", "st_rad", "st_mass", "st_met", "st_lum",
    "sy_dist"
]

df = df[columns_needed]
df.head()
```

```python
df.isnull().sum()
```

```python

threshold = len(df) * 0.4
df = df.dropna(thresh=threshold, axis=1)
df.shape
```

```python
df = df.fillna(df.median(numeric_only=True))
```

```python
df.isnull().sum()
```

```python
df.to_csv("exoplanet_clean.csv", index=False)
```
