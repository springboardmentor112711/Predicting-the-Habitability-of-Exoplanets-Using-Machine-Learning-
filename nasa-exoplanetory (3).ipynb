# I(InterQuartile)QR method for the outlier detection 

```python
cols=['planet_size','planet_temp','star_temp','star_energy','star_gravity','star_size','brightness']
for col in cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    df=df[(df[col]>=Q1-1.5*IQR)&(df[col]<=Q3+1.5*IQR)]

```

```python
X = df[['planet_temp_z', 'planet_size_z', 'star_temp_z', 'star_energy_z']]
y = df['final_result_CONFIRMED']
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=42,
    use_label_encoder=False
)
```

***Regularization***:Regularization is a technique in machine learning used to prevent overfitting by penalizing complex models, so the model learns only the important patterns and performs well on new data.

```python
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 4],
    'min_child_weight': [3, 5],
    'subsample': [0.8],
    'colsample_bytree': [0.8],
    'reg_alpha': [0.1, 0.5],     # L1 regularization
    'reg_lambda': [1.0, 2.0]     # L2 regularization
}
```

**Cross validation is a technique used to check how well a machine learning model will work on new, unseen data by testing it on different parts of the dataset.**

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid.fit(X, y)
```

```python
print("Best CV Accuracy:", grid.best_score_)
print("Best Parameters:", grid.best_params_)
```

```python
best_xgb = grid.best_estimator_
best_xgb.fit(X, y)
```
